```python
import seaborn as sns
import matplotlib.pyplot as plt
# If you're using a Jupyter notebook, use the magic function %matplotlib inline to display plots inline
%matplotlib inline  
# Load the tips dataset
dataset = sns.load_dataset('tips')
# Show the first five entries of the dataset
print(dataset.head())
numeric_dataset=dataset.select_dtypes(include=['float64','int64'])
# Compute the correlation between the different parameters
tc =numeric_dataset.corr()  # correlation asign numberic value only.
# Plot a heatmap of the correlation data
sns.heatmap(tc, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Load the tips dataset
tips = sns.load_dataset('tips')
# Inspect the first few rows to see the columns and data types
print(tips.head())
# Check data types to find non-numeric columns
print(tips.dtypes)
# Option 1: Drop non-numeric columns (like 'sex', 'smoker', 'day', 'time')
numeric_tips = tips.select_dtypes(include=['number'])
# Option 2: Alternatively, convert categorical columns to numeric (for example, 'smoker' and 'sex')
# You can convert binary categorical variables to numeric (Yes/No → 1/0)
tips['smoker'] = tips['smoker'].map({'Yes': 1, 'No': 0})
tips['sex'] = tips['sex'].map({'Male': 1, 'Female': 0})
# Now calculate the correlation matrix on numeric columns
correlation_matrix = numeric_tips.corr()
# Create the heatmap
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=True)
# Show the plot
plt.show()
```
